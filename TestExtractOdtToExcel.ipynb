```python
from odf.opendocument import load
from odf.text import P
from odf.opendocument import OpenDocumentSpreadsheet
from odf.table import Table, TableRow, TableCell
from odf.text import P as TableP
import pandas as pd
from openpyxl import Workbook


class MINERAL:
    def __init__(self, name, color=None, formula=None, formulaWeight=None, otherSpecies=None, discovery=None, system=None, 
                 magnetism=None, crystal_class=None, radioactivity=None, mohs=None, density=None, fusion_temperature=None,
                 macle=None, fluorescence=None, cleavage=None, habitus=None, glow=None, pleochroism=None, chemicalBehaviour=None,
                 transparence=None, gite=None, associatedMinerals=None, gisement=None, use=None, synonyme=None, topotype=None, infoSup=None):
        self.name = name
        self.color = color
        self.formula = formula
        self.formulaWeight = formulaWeight
        self.otherSpecies = otherSpecies
        self.discovery = discovery
        self.system = system
        self.magnetism = magnetism
        self.crystal_class = crystal_class
        self.radioactivity = radioactivity
        self.mohs = mohs
        self.density = density
        self.fusion_temperature = fusion_temperature
        self.macle = macle
        self.fluorescence = fluorescence
        self.cleavage = cleavage
        self.habitus = habitus
        self.glow = glow
        self.pleochroism = pleochroism
        self.chemicalBehaviour = chemicalBehaviour
        self.transparence =  transparence
        self.gite = gite
        self.associatedMinerals = associatedMinerals
        self.gisement = gisement
        self.use = use
        self.synonyme = synonyme
        self.topotype = topotype
        self.infoSup = infoSup

    def to_dict(self):
        return {
            'Name': self.name,
            'Color': self.color,
            'Formula': self.formula,
            'FormulaWeight': self.formulaWeight,
            'OtherSpecies': self.otherSpecies,
            'Discovery': self.discovery,
            'System': self.system,
            'Magnetism': self.magnetism,
            'Crystal Class': self.crystal_class,
            'Radioactivity': self.radioactivity,
            'Mohs': self.mohs,
            'Density': self.density,
            'Fusion_Temperature': self.fusion_temperature,
            'Macle': self.macle,
            'Fluorescence': self.fluorescence,
            'Cleavage': self.cleavage,
            'Habitus': self.habitus,
            'Glow': self.glow,
            'Pleochroism': self.pleochroism,
            'ChemicalBehaviour': self.chemicalBehaviour,
            'Transparence': self.transparence,
            'Gite': self.gite,
            'associatedMinerals': self.associatedMinerals,
            'Gisement': self.gisement,
            'Use': self.use,
            'Synonyme': self.synonyme,
            'Topotype': self.topotype,
            'InfoSup': self.infoSup
        }

def get_paragraph_text(paragraph):
    texts = []
    for node in paragraph.childNodes:
        if node.nodeType == node.TEXT_NODE:
            texts.append(node.data)
        elif node.nodeType == node.ELEMENT_NODE and node.tagName == 'text:span':
            texts.append(node.firstChild.data)
    return ''.join(texts).strip()

# Charger le document ODT
doc = load('MineralTestExtract.odt')

# Parcourir les paragraphes du document pour extraire les informations
current_mineral = None
Mineral_data = []
Index, i=[], -1

#print(doc.getElementsByType(P))

for paragraph in doc.getElementsByType(P):
    i+=1
    text = get_paragraph_text(paragraph)
    #print(text)
    #print('\n')
    key_value = text.split(" ")
    #print(key_value)
    if len(key_value)==1:
        if key_value[0].find(":")==-1 and key_value[0]!="":
            Index.append(i)
Index.append(len(doc.getElementsByType(P)))      

for i in range(len(Index)-1): 
    mineral_data = MINERAL(i)
    text = get_paragraph_text(doc.getElementsByType(P)[Index[i]])
    key_value = text.split(" ")
    mineral_data.name=key_value[0]
    for i in range(Index[i]+1,Index[i+1]):
        text = get_paragraph_text(doc.getElementsByType(P)[i])
        key_value = text.split(": ")
        #print(key_value)
        if key_value[0].find("Couleur")!=-1:
            mineral_data.color=key_value[1:][0]
        elif key_value[0].find("Formule")!=-1:
            mineral_data.formula=key_value[1:][0]
        elif key_value[0].find("Masse formulaire")!=-1:
            mineral_data.formulaWeight=key_value[1:][0]
        elif key_value[0].find("Autres espèces possiblement présentes")!=-1:
            mineral_data.otherSpecies=key_value[1:][0]
        elif key_value[0].find("Découverte")!=-1:
            mineral_data.discovery=key_value[1:][0]
        elif key_value[0].find("Système cristallin")!=-1:
            mineral_data.system=key_value[1:][0]
        elif key_value[0].find("Magnétisme")!=-1:
            mineral_data.magnetism=key_value[1:][0]
        elif key_value[0].find("Classe cristalline")!=-1:
            mineral_data.crystal_class=key_value[1:][0]
        elif key_value[0].find("Radioactivité")!=-1:
            mineral_data.radioactivity=key_value[1:][0]
        elif key_value[0].find("Echelle de Mohs")!=-1:
            mineral_data.mohs=key_value[1:][0]
        elif key_value[0].find("Densité")!=-1:
            mineral_data.density=key_value[1:][0]
        elif key_value[0].find("Température de fusion")!=-1:
            mineral_data.fusion_temperature=key_value[1:][0]
        elif key_value[0].find("Macle")!=-1:
            mineral_data.macle=key_value[1:][0]
        elif key_value[0].find("Fluorescence")!=-1:
            mineral_data.fluorescence=key_value[1:][0]
        elif key_value[0].find("Clivage")!=-1:
            mineral_data.cleavage=key_value[1:][0]
        elif key_value[0].find("Habitus")!=-1:
            mineral_data.habitus=key_value[1:][0]
        elif key_value[0].find("Eclat")!=-1:
            mineral_data.glow=key_value[1:][0]
        elif key_value[0].find("Pléochroïsme")!=-1:
            mineral_data.pleochroism=key_value[1:][0]
        elif key_value[0].find("Comportement chimique")!=-1:
            mineral_data.chemicalBehaviour=key_value[1:][0]
        elif key_value[0].find("Transparence")!=-1:
            mineral_data.transparence=key_value[1:][0]
        elif key_value[0].find("Gîte")!=-1:
            mineral_data.gite=key_value[1:][0]
        elif key_value[0].find("associée aux minéraux")!=-1:
            mineral_data.associatedMinerals=key_value[1:][0]
        elif key_value[0].find("Gisement")!=-1:
            mineral_data.gisement=key_value[1:][0]
        elif key_value[0].find("Utilisation")!=-1:
            mineral_data.use=key_value[1:][0]
        elif key_value[0].find("Synonyme")!=-1:
            mineral_data.synonyme=key_value[1:][0]
        elif key_value[0].find("Topotype")!=-1:
            mineral_data.topotype=key_value[1:][0]
        elif key_value[0].find("Info Sup")!=-1:
            mineral_data.infoSup=key_value[1:][0]
   
    Mineral_data.append(mineral_data)
        
print('Mineral_data')
print(Mineral_data)
print(Mineral_data[0].name)
print(Mineral_data[1].name)
print(Mineral_data[2].name)


# Créer un nouveau classeur Excel
wb = Workbook()

# Sélectionner la feuille active (la première feuille par défaut)
ws = wb.active

# Nommer la feuille (optionnel)
ws.title = "Minerals Data"

# Ajouter des en-têtes de colonnes
headers = ["Name", "Formula", "Color", "FormulaWeight"]
ws.append(headers)

# Ajouter les données des instances de MINERAL
for mineral in Mineral_data:
    ws.append([mineral.name, mineral.formula, mineral.color, mineral.formulaWeight])

# Enregistrer le fichier Excel
wb.save("minerals_data.xlsx")
```
